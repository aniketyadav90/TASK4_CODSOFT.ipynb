```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ads[['TV']], ads[['Sales']], test_size=0.4, random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
res=model.predict(x_test)
print(res)
```

```python
model.coef_
```

```python
model.intercept_
```
